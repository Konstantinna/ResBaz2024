```python
df.head()
```

#### Remove unwanted columns by name or parts of the column name


```python
#df.drop(columns=['name'])
# Create a boolean array indicating which columns contain the string "Unnamed"
cols_to_drop = df.columns[df.columns.str.contains('Unnamed')]

# Drop the columns containing the string "Unnamed"
df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
df
```

## Visualising your data

Note: Some libraries that are common for classical Jupyter Notebooks in Jupyter Lab might get errors when you try to run things in Lite
_Altair_ is a handy and interactive one that you can use here in the browser-based version.

```python
import micropip
await micropip.install("altair")
import altair as alt
```

```python
#%pip install summarytools
%pip install Jinja2
from summarytools import dfSummary
```

```python
dfSummary(df)
```

```python
#%pip install klib
import klib
```

```python
dfcorr = df[['Age','ArtAbility','totalvviq','totalosiq']]
```

```python
klib.corr_plot(dfcorr)
```

```python
# Altair in Lite: let's create a bar chart for the biggest bands (more than 100 fans)
alt.Chart(df).mark_bar().encode(
    alt.X("totalvviq", bin=True),
    y='count()',
)
```

```python
# Let's look at some relationships between variables
alt.Chart(df).mark_circle(size=60).encode(
    x='Age',
    y='totalvviq',
    color='aphantasia',
).interactive()
```

```python
# Let's create a scatterplot with additional info incorporated as colour and tooltip

alt.Chart(df).mark_circle(size=60).encode(
    x='totalvviq',
    y='totalosiq',
    color='Country',
    tooltip=['MostDifficult']
).interactive()
```
